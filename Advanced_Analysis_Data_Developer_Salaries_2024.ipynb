# Comprehensive Analysis of Data Developer Salaries and Employment Attributes (2024)

This notebook provides a comprehensive analysis of the 2024 dataset on data developer salaries and employment attributes.

## Importing required Libraries and Dataset

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from scipy.stats import pearsonr

df = pd.read_csv('/data/Dataset salary 2024.csv')
df.replace([np.inf, -np.inf], np.nan, inplace=True)

df.head()
```

## Basic Dataset Overview

```python
df.describe(include='all')
```

## Salary Distribution

```python
plt.figure(figsize=(12, 6))
sns.histplot(df['salary_in_usd'].dropna().values, bins=30, kde=True)
plt.title('Distribution of Salaries (in USD)')
plt.xlabel('Salary (USD)')
plt.ylabel('Frequency')
plt.show()
```

## Experience Level vs Salary

```python
plt.figure(figsize=(12, 6))
sns.boxplot(x='experience_level', y='salary_in_usd', data=df)
plt.title('Salary by Experience Level')
plt.xlabel('Experience Level')
plt.ylabel('Salary (USD)')
plt.show()
```

## Employment Type vs Salary

```python
plt.figure(figsize=(12, 6))
sns.boxplot(x='employment_type', y='salary_in_usd', data=df)
plt.title('Salary by Employment Type')
plt.xlabel('Employment Type')
plt.ylabel('Salary (USD)')
plt.show()
```

## Job Title vs Salary

```python
top_jobs = df.groupby('job_title')['salary_in_usd'].mean().sort_values(ascending=False).head(10)

plt.figure(figsize=(12, 6))
sns.barplot(x=top_jobs.values, y=top_jobs.index)
plt.title('Top 10 Job Titles by Average Salary')
plt.xlabel('Average Salary (USD)')
plt.ylabel('Job Title')
plt.show()
```

## Geographic Analysis

```python
top_locations = df.groupby('company_location')['salary_in_usd'].mean().sort_values(ascending=False).head(10)

plt.figure(figsize=(12, 6))
sns.barplot(x=top_locations.values, y=top_locations.index)
plt.title('Top 10 Company Locations by Average Salary')
plt.xlabel('Average Salary (USD)')
plt.ylabel('Company Location')
plt.show()
```

## Remote Work Analysis

```python
plt.figure(figsize=(12, 6))
sns.boxplot(x='remote_ratio', y='salary_in_usd', data=df)
plt.title('Salary by Remote Work Ratio')
plt.xlabel('Remote Work Ratio (%)')
plt.ylabel('Salary (USD)')
plt.show()
```

## Company Size Analysis

```python
plt.figure(figsize=(12, 6))
sns.boxplot(x='company_size', y='salary_in_usd', data=df)
plt.title('Salary by Company Size')
plt.xlabel('Company Size')
plt.ylabel('Salary (USD)')
plt.show()
```

## Correlation Analysis

```python
numeric_df = df.select_dtypes(include=['float64', 'int64'])
corr = numeric_df.corr()

plt.figure(figsize=(12, 8))
sns.heatmap(corr, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Correlation Matrix')
plt.show()
```

## Summary and Insights

This notebook provides a detailed analysis of the 2024 dataset on data developer salaries and employment attributes. The analysis includes salary distribution, experience level, employment type, job title, geographic analysis, remote work analysis, company size analysis, and correlation analysis.
